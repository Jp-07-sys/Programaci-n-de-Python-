5.- Un objeto de masa m=2 kg se encuentra a una altura h=10 cm y tiene una velocidad de v=3 m/s. Calcular:

a) La energía cinética (Ec),

b) La energía potencial gravitacional (Ep),

c) La energía mecánica total (Em).

Considere g=9.8m/s^2. Revise las unidades.

```python
#Datos del problema
m=2 # kg
h=10 # cm
v=3 # m/s
g=9.8 #m/s^2
#Cálculo dela energía cinetica
Ec=1/2 * m * v**2
print("Energía cinética",Ec,"J")
#Cálculo de energía potencial gravitacional
Ep=m * g * h
print("Energía potencial gravitacional",Ep,"J")
#Cálculo de energía mecánica total
Em= Ec + Ep
print("Energía mecánica total",Em,"J")
```

 7.- Tipo de triángulo.

Solicite al usuario que ingrese los valores de los lados de un triángulo (enteros positivos) y realice lo siguiente:

a).- Valide que formen un triángulo (desigualdad triangular).

b).- Si forman un triángulo, indica si es equilátero, isosceles o escaleno.

```python
#Valores de los lados del triangúlo
a= int(input("Ingrese el valor del primer lado a : "))
b= int(input("Ingrese el valor del segundo lado b : "))
c= int(input("Ingrese el valor del tercer lado c : "))
#Desigualdad triangular
if a + b > c and  a + c > b and b + c > a:
  print("El triángulo es válido")
else:
  print("El triángulo no es válido")
#Tipo de triángulo
if a == b == c:
  print("El triángulo es equilátero")
if a == b or a == c or b == c :
  print("El triángulo es isósceles")
if a!=b!=c:
  print("El triángulo es escaleno")
```
