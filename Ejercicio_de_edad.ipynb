```python
#Edad del usuario
edad = int(input("Ingresa tu edad: "))
if edad >= 0 and edad <= 11:
  print("NiÃ±o")
elif edad >= 12 and edad <= 17:
  print("Adolescente")
elif edad >= 18 and edad <= 29:
  print("Joven")
elif edad >= 30 and edad <= 64:
  print("Adulto")
elif edad >= 65:
  print("Adulto mayor")
else:
  print("Edad incorrecta, debe tener un valor positivo")
```
