```python
#Programe una función que calcule el área de un triángulo utilizando la fórmula de Herón
#Variables globales
a= int(input("Ingrese el primer lado: "))
b= int(input("Ingrese el segundo lado: "))
c= int(input("Ingrese el tercer lado: "))
# Que los valores del triángulo sean positivos. El programa debe repetirse hasta que los tres valores sean positivos.
while not a>0 or not b>0 or not c>0:
  print("Los valores ingresados deben ser positivos.")
  a= int(input("Ingrese el primer lado: "))
  b= int(input("Ingrese el segundo lado: "))
  c= int(input("Ingrese el tercer lado: "))
#Desigualdad triangular
if a <=0 and b <=0 and c <=0:
  print("Los valores ingresados no forman un triángulo")
elif a + b > c and  a + c > b and b + c > a:
 print("El triángulo es válido")
elif  a + b <= c  or a + c <= b or b + c <= a:
  print("El triángulo no es válido")
#Fórmulas
def area(a,b,c):
  s= (a+b+c)/2 #Variable locales
  area= (s*(s-a)*(s-b)*(s-c))**1/2 #Variable locales
  return area

print("El área del triángulo es: ", area(a,b,c))
```
