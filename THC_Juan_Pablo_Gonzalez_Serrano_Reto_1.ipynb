```python
#Genere una función que le permita jugar piedra, papel o tijera con la computadora, teniendo en cuenta tres posibles resultados: victoria, empate o derrota.
import random
def jugar_piedra_papel_tijera():
    opciones = ['piedra', 'papel', 'tijera']
    computadora = random.choice(opciones)
    jugador = input("Elige piedra, papel o tijera: ")

    if jugador not in opciones:
        print("Opción no válida. Por favor elige piedra, papel o tijera.")
        return

    print(f"La computadora eligió: {computadora}")

    if jugador == computadora:
        resultado = "Empate"
    elif (jugador == 'piedra' and computadora == 'tijera') or \
         (jugador == 'papel' and computadora == 'piedra') or \
         (jugador == 'tijera' and computadora == 'papel'):
        resultado = "Victoria"
    else:
        resultado = "Derrota"
    print(resultado)

jugar_piedra_papel_tijera()
```
