# **Ejercicio 1.- Ingreso a una boveda.**

Pedro y Jes√∫s son dos hermanos, hijos de Manuel, y quienes tienen acceso a una b√≥veda en un exclusivo banco. El due√±o original de esta b√≥veda es su padre, y en su testamento ha especificado las siguientes condiciones para que el banco le permita acceder a sus hijos al contenido de dicha b√≥veda. Las condiciones son las siguientes:

a. Juan debe ser mayor de 35 a√±os, o Pedro debe ser menor de 50 a√±os.

b. La suma de sus edades debe ser un n√∫mero par.

c. Manuel debe tener m√≠nimo 1 nieto.

Construya una funci√≥n que lea la edad de Juan, la edad de Pedro, el n√∫mero de nietos y calcule, usando condicionales, si tienen acceso (SI) o no (NO) a la b√≥veda.

```python
juan = int(input("Ingrese la edad de Juan: "))
pedro = int(input("Ingrese la edad de Pedro: "))
nietos = int(input("Ingrese el n√∫mero de nietos: "))

def edad (juan,pedro,nietos):

    condicion_a = juan > 35 or pedro < 50

    condicion_b = juan + pedro

    condicion_c = nietos >= 1

    if condicion_a and condicion_b and condicion_c:
        return "SI"
    else:
        return "NO"

resultado: edad(juan, pedro, nietos)

print(f"Tienen acceso a la b√≥veda: {resultado}")
```

**Ejercicio 3.- N√∫mero primo.**

Con base en la definici√≥n de n√∫mero primo, escriba una funci√≥n que reciba un n√∫mero entero e imprima 'True' si lo es o de lo contrario imprima 'False'. Tenga en cuenta que si el n√∫mero es menor o igual a 1, se retorna como resultado "Error en el N√∫mero" y debe volver a preguntar hasta que la entrada sea valida.

```python
num=int(input("Ingrese un numero: "))
# Validar el n√∫mero entero
while num <= 1:
    print("Error en el n√∫mero, debe ser mayor que 1.")
    num = int(input("Ingrese un n√∫mero: "))

# Ver si es primo
primo = True  #asumimos que si es primo
for i in range(2, num):   #si tiene alg√∫n divisor exacto
    if num % i == 0:
        primo = False
        break

if primo:
    print("True (es primo)")
else:
    print("False (no es primo)")
```

**Ejercicio 4.- N√∫meros mayores a 100.**

Escriba una funci√≥n que lea un n√∫mero entero (N) mayor que 100 y valide:

 1. Si N es menor de 100 retorne: "Error en el n√∫mero" y vuelva a preguntar la entrada.
 2. Si N termina en 4, 7 √≥ 9; diga si es divisible por 3.
 3. Si N termina en 1 √≥ 5, divida por 10 y retorne el valor entero,
 4. En otro caso divida el n√∫mero entre 3 y retorne su residuo.

```python
num=int(input("Ingrese un numero: "))

while num < 100:
    print("Error en el numero")
    num=int(input("Ingrese un numero: "))

def num_mayor(num):

    if num % 10 == 4 or num % 10 == 7 or num % 10 == 9:
      return "Es divisible entre 3"
    elif num % 10 == 1 or num % 10 == 5:
      return num // 10
    else:
      return num % 3

num_mayor(num)
```

**Ejercicio 5.- MCD y MCM.**

Dados dos n√∫meros a y b, determine:

* El m√°ximo com√∫n divisor.
* El m√≠nimo com√∫n m√∫ltiplo.

Almacene los resultados en un diccionario e imprima en pantalla el contenido del diccionario.

```python
import math

a=int(input("Ingrese un numero: "))
b=int(input("Ingrese un numero: "))

mcd= lambda a,b: abs(a*b) // math.gcd(a,b)
mcm= lambda a,b: abs(a*b) // math.lcm(a,b)

print("mcd:",mcd(a,b))
print("mcm:",mcm(a,b))

#Almacenar los resultados en un diccionario
resultados = {
    "MCD": mcd(a,b),
    "MCM": mcm(a,b)
}
print(f"Resultados:{resultados}")
```

**Ejercicio 6.- Filtrado de datos.**

Escriba una funci√≥n que reciba una lista de datos (la de abajo) y devuelva listas con los datos filtrados. Este programa debe separar los datos en int, str, bool, float.

**Hint**: use la funci√≥n append para insertar datos en las listas.

```python
datos = [25, "Matem√°ticas", 3.141592, True, 8, False, "Nombres", 2.718, "UNAM", 0,
         9.81, "CdMx", 7e12, 12.5, "2025", True, "poblaci√≥n", 0.01, 42, False]

def filtrar_datos(lista):
    enteros = []
    flotantes = []
    cadenas = []
    booleanos = []

    for dato in lista: #isintance es para comprobar el dato que quieres
        if isinstance(dato, int) and not isinstance(dato, bool):  # Evita que True/False se metan como enteros
            enteros.append(dato)
        elif isinstance(dato, float):
            flotantes.append(dato)
        elif isinstance(dato, str):
            cadenas.append(dato)
        elif isinstance(dato, bool):
            booleanos.append(dato)

    return enteros, flotantes, cadenas, booleanos

# Llamamos a la funci√≥n
enteros, flotantes, cadenas, booleanos = filtrar_datos(datos)

# Mostramos los resultados
print("Enteros:", enteros)
print("Flotantes:", flotantes)
print("Cadenas:", cadenas)
print("Booleanos:", booleanos)
```

**Ejercicio 7.- C√°lculo de la distancia entre dos puntos mediante la f√≥rmula de Haversine.**

La f√≥rmula de Haversine permite calcular la distancia entre dos puntos sobre la superficie de la Tierra, a partir de sus coordenadas geogr√°ficas. La f√≥rmula est√° dada por la siguiente expresi√≥n:

\begin{equation}
d = 2R  \arcsin \left( \sqrt{\mathrm{haversin}(\varphi_2 - \varphi_1) + \cos(\varphi_1)\cos(\varphi_2)\mathrm{haversin}(\lambda_2 - \lambda_1)} \right)
\end{equation}

donde:

d = distancia entre dos puntos (en km),

R = radio de la Tierra (ùëÖ‚âà6378.1),

ùúë1, ùúë2 = latitudes de los dos puntos.

ùúÜ1, ùúÜ2 = longitudes de los dos puntos (en radianes).

La funci√≥n de Haversine se define como:

 \begin{equation}
\mathrm{haversin}(\alpha) = \sin^2 \left(\frac{\alpha}{2}\right)
\end{equation}

Para convertir de grados a radianes se utiliza la relaci√≥n:

\begin{equation}
\text{radianes} = \text{grados} \times \frac{\pi}{180}
\end{equation}

Construya una funci√≥n que permita calcular la distancia entre dos pares de coordenadas. Utiliza la librer√≠a math para las funciones trigonom√©tricas y de ra√≠z cuadrada.

**Ejemplo de prueba.** Coordenadas de Paris y Roma por latitud y longitud: Par√≠s (48.9, 2.4) y Roma (41.9, 12.5). Resultado: distancia=1107 km.

Proponga 3 pares de coordenadas. Al menos uno debe ser respecto a la Ciudad de M√©xico (**Latitud=19.43, Longitud=-99.13**).

```python
import math

# Radio de la Tierra en km
R = 6378.1

# Funci√≥n para convertir grados a radianes
def grados_a_radianes(grados):
    return grados * math.pi / 180

# Funci√≥n de Haversine
def haversine(lat1, lon1, lat2, lon2):
    # Convertir coordenadas a radianes
    lat1, lon1, lat2, lon2 = map(grados_a_radianes, [lat1, lon1, lat2, lon2])

    # Diferencias
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # F√≥rmula de Haversine
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    d = R * c
    return d

# Ejemplo : Par√≠s (48.9, 2.4) y Roma (41.9, 12.5)
distancia_paris_roma = haversine(48.9, 2.4, 41.9, 12.5)
print(f"Distancia entre Par√≠s y Roma: {distancia_paris_roma:.0f} km")

# Tres pares de coordenadas (uno con CDMX)
coordenadas = [
    ((19.43, -99.13), (34.05, -118.25)),  # CDMX - Los √Ångeles
    ((19.43, -99.13), (40.71, -74.00)),   # CDMX - Nueva York
    ((48.9, 2.4), (35.68, 139.69))        # Par√≠s - Tokio
]

# Calcular y mostrar las distancias
for (lat1, lon1), (lat2, lon2) in coordenadas:
    d = haversine(lat1, lon1, lat2, lon2)
    print(f"Distancia entre ({lat1}, {lon1}) y ({lat2}, {lon2}): {d:.0f} km")
```

**Ejercicio 8.- La serie de Fibonacci.**

Realice un programa que calcule la serie de Fibonacci y la imprima en pantalla, tenga en cuenta las restricciones de la variable de entrada.

```python
#Cada numero es la suma de los dos anteriores
def fibonacci(n):
    serie = [0, 1]  # primeros dos t√©rminos
    for i in range(2, n):
        serie.append(serie[i-1] + serie[i-2])
    return serie[:n]
print(fibonacci(10))
```

**Ejercicio 9.- Mini-cat√°logo de sismos.**

En este ejercicio se muestran 4 listas que representan un conjunto de sismos con la siguiente informaci√≥n (fecha, zona, magnitud y profundidad). Hacer un programa que haga lo siguiente:

1.- Convierte la lista de magnitues de str a float.

2.- Los nombres de los estados deben iniciar con may√∫scula (utilice una funci√≥n lambda).

3.- Crea una lista llamada intensidad (con la misma longitud de la lista de magnitudes) y coloca etiquetas de acuerdo al siguiente criterio:

*   Magnitud < 3.5 "Muy Leve"
*   3.5 <= Magnitud < 4.5 "Leve"
*   4.5 <= Magnitud < 3.5 "Moderado"
*   Magnitud > 5.5 "Fuerte"

4.- Con un ciclo For (y condicionales si es necesario), genera un diccionario "conteo_por_zona" con el n√∫mero de eventos por zona.

5.- Ordena los eventos por magnitud descendente y muestra los 3 sismos m√°s grandes con el siguiente formato:

  YYYY-MM-DD  |  Zona (Estado)  |  Magnitud   |  Profundidad (km).

**Hint:** Recuerde las funciones predefinidas de python map, filter, len, entre otras.


```python
fechas = ["2025-03-01","2025-03-02","2025-03-03","2025-03-04","2025-03-05",
    "2025-03-06","2025-03-07","2025-03-08","2025-03-09","2025-03-10"]

zonas = ["guerrero","oaxaca","chiapas","michoac√°n","guerrero",
    "oaxaca","chiapas","guerrero","oaxaca","chiapas"]

magnitudes = ["3.8","4.2","5.0","2.9","4.8","3.5","4.1","5.3","4.0","3.0"]

prof_km = [15, 80, 12, 5, 35, 120, 60, 25, 10, 90] #Profundidad en kil√≥metros.


# Convertir magnitudes a float
magnitudes = list(map(float, magnitudes))

# Capitalizar nombres de los estados (lambda)
zonas = list(map(lambda z: z.capitalize(), zonas))

# Crear lista de intensidades seg√∫n la magnitud
def clasificar_intensidad(m):
    if m < 3.5:
        return "Muy leve"
    elif 3.5 <= m < 4.5:
        return "Leve"
    elif 4.5 <= m < 5.5:
        return "Moderado"
    else:
        return "Fuerte"

intensidades = list(map(clasificar_intensidad, magnitudes))

# Crear diccionario con conteo de eventos por zona
conteo_por_zona = {}
for zona in zonas:
    conteo_por_zona[zona] = conteo_por_zona.get(zona, 0) + 1

# Ordenar por magnitud descendente y mostrar los 3 sismos m√°s grandes
sismos = list(zip(fechas, zonas, magnitudes, prof_km, intensidades))
sismos_ordenados = sorted(sismos, key=lambda x: x[2], reverse=True)

print("Mini Cat√°logo de Sismos")
for sismo in sismos_ordenados[:3]:
    print(f"{sismo[0]} | {sismo[1]} | {sismo[2]} | {sismo[3]} km | {sismo[4]}")

print("Conteo de eventos por zona")
for zona, conteo in conteo_por_zona.items():
    print(f"{zona}: {conteo}")
```

**Ejercicio 10.- Operaciones vectoriales.**

Dadas las siguientes listas:

a = [2, -1, 4]

b = [4, -2, 8]

c = [-5, 2, 1]

Calcule el producto escalar:

1) ab

2) bc

3) ca

Calcular el productor vectorial:

1) a **x** b

2) b **x** c

3) a **x** c

Utilice las listas de inicio para probar su c√≥digo. Posteriormente, genere una funci√≥n que reciba como entrada las listas (vectores) y haga las operaciones vectoriales.

Indique si algunos de los vectores son paralelos o perpendiculares.

```python
a = [2, -1, 4]
b = [4, -2, 8]
c = [-5, 2, 1]

#Producto escalar
def producto_escalar(v1, v2):
    return sum(v1[i] * v2[i] for i in range(len(v1)))

#Producto vectorial
def producto_vectorial(v1, v2):
    return [
        v1[1]*v2[2] - v1[2]*v2[1],
        v1[2]*v2[0] - v1[0]*v2[2],
        v1[0]*v2[1] - v1[1]*v2[0]
    ]

#C√°lculos de productos escalares
ab = producto_escalar(a, b)
bc = producto_escalar(b, c)
ca = producto_escalar(c, a)

print("Producto escalar a¬∑b =", ab)
print("Producto escalar b¬∑c =", bc)
print("Producto escalar c¬∑a =", ca)

#C√°lculos de productos vectoriales
axb = producto_vectorial(a, b)
bxc = producto_vectorial(b, c)
axc = producto_vectorial(a, c)

print("Producto vectorial a √ó b =", axb)
print("Producto vectorial b √ó c =", bxc)
print("Producto vectorial a √ó c =", axc)

#Verificar paralelismo y perpendicularidad

def perpendicularidad(v1, v2):
    return producto_escalar(v1, v2) == 0

def paralelismo(v1, v2):
    razones = []
    for i in range(len(v1)):
        if v2[i] != 0:
            razones.append(v1[i] / v2[i])
    return all(r == razones[0] for r in razones)

print("¬øa y b son perpendiculares?", perpendicularidad(a, b))
print("¬øb y c son perpendiculares?", perpendicularidad(b, c))
print("¬øa y c son perpendiculares?", perpendicularidad(a, c))

print()
print("¬øa y b son paralelos?", paralelismo(a, b))
print("¬øb y c son paralelos?", paralelismo(b, c))
print("¬øa y c son paralelos?", paralelismo(a, c))
```
