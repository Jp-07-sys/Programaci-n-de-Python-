Calcule las sumas superiores e inferiores de las siguientes funciones:


*   f(x) = x^3 en [0, 2]
*   g(x) = 4 − x^2 en [−1, 2]

```python
import numpy as np

# Definimos la función f(x) = x^3
def f(x):
    return x**3

# Definimos la función g(x) = 4 - x^2
def g(x):
    return 4 - x**2

# Definimos una función que calcule suma inferior y superior
def sumas(funcion, a, b, n):

    # Calculamos el ancho de cada subintervalo
    dx = (b - a) / n

    # Generamos los puntos de la partición
    particion = np.linspace(a, b, n + 1)

    # Inicializamos suma inferior
    suma_inf = 0

    # Inicializamos suma superior
    suma_sup = 0

    # Recorremos cada subintervalo
    for i in range(n):
        x_izq = particion[i]
        x_der = particion[i + 1]

        # Evaluamos la función en ambos extremos
        valores = [funcion(x_izq), funcion(x_der)]

        # Sumamos el mínimo por dx para la suma inferior
        suma_inf += min(valores) * dx

        # Sumamos el máximo por dx para la suma superior
        suma_sup += max(valores) * dx

    # Retornamos ambas sumas
    return suma_inf, suma_sup

# Lista de valores de n para observar el comportamiento
valores_n = [5, 10, 50, 100]

# ---- Cálculo para f(x) = x^3 en [0,2] ----
print("Resultados para f(x) = x^3 en [0,2]\n")

for n in valores_n:
    # Calculamos sumas inferior y superior
    inf, sup = sumas(f, 0, 2, n)

    # Imprimimos resultados
    print(f"n = {n}")
    print(f"Suma inferior: {"%.3f" % inf}")
    print(f"Suma superior: {"%.3f" % sup} ")

# Valor exacto de la integral de x^3 en [0,2]
valor_exacto_f = (2**4) / 4
print("Valor exacto de la integral:", valor_exacto_f)

print("\n-----------------------------------\n")

# ---- Cálculo para g(x) = 4 - x^2 en [-1,2] ----
print("Resultados para g(x) = 4 - x^2 en [-1,2]\n")

for n in valores_n:
    # Calculamos sumas inferior y superior
    inf, sup = sumas(g, -1, 2, n)

    # Imprimimos resultados
    print(f"n = {n}")
    print(f"Suma inferior: {"%.3f" % inf}")
    print(f"Suma superior: {"%.3f" % sup}")

# Valor exacto de la integral de 4 - x^2 en [-1,2]
valor_exacto_g = (4*2 - (2**3)/3) - (4*(-1) - ((-1)**3)/3)
print("Valor exacto de la integral:", valor_exacto_g)
```
